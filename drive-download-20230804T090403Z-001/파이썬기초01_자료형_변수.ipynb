```python
def create_dictioanry():
  dictionary_list = []
  while True :
    Key = input('key 값 입력하세용>> ')
    user_prompt = {}
    if Key == 'q':
      break
    Value = input('value 값 입력하세용 >> ')
    user_prompt[Key] = Value
    dictionary_list.append(user_prompt)
  return dictionary_list
```
