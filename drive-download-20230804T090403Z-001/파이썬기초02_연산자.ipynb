## 멤버 연산자
| in | 왼쪽항이 오른쪽 리스트 안에 들어 있으면 참 |
| --- | --- |
| not in | 왼쪽항이 오른쪽 리스트 안에 안들어 있으면 참 |



```python
mem = [1,2,3,4,5]
```

```python
1 in mem
```

```python
7 in mem
```

```python
6 not in mem
```
