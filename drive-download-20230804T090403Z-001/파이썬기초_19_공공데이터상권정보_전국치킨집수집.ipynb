```python
import os
import sys
import urllib.request
import time
import datetime
from urllib.parse import urlencode, unquote
import getpass
import json
import pandas as pd 
```

```python
def get_request_url(url):
  req = urllib.request.Request(url)
  try:
    response = urllib.request.urlopen(req)
    if response.getcode() == 200:
      return response.json
  except Exception as e:
    print(e)
    print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
  return None


def get_urls_open(ServiceKey,pageNo,numOfRows,divId,key,Type):
    service_url = "https://apis.data.go.kr/B553077/api/open/sdsc2/"
    parameters = {
        "serviceKey": ServiceKey,  # 인증키
        "pageNo": pageNo,
        "numOfRows": numOfRows,
        "divId": divId,
        "key": key,
        "type": Type
    }

    url = service_url + "storeListInUpjong?" + urlencode(parameters)

    response_decode = get_request_url(url)  # [CODE 1]
    
    if(response_decode == None):
        return None
    else :
        data = json.loads(response_decode)
        with open('치킨집정보.json','w',encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        return data
```

```python
#전체 페이지 저장 할 리스트 
# While True 반복문 
# 전체데이터 / 1000 : 전체페이지수 
# 페이지 증가 pageNo 
# page > 전체 페이지 수 = 반복문 종료

sevice_key = getpass.getpass(' 서비스키 입력 >> ')
result = get_urls_open(sevice_key,1,1000,'indsSclsCd','I21006','json')
print(result)
```
