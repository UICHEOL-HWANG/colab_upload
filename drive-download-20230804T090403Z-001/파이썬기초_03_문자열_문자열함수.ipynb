## 문자열 함수
* 문자열을 조작하는 함수
* .split().join().strip().replace()

* split() : 기준 문자를 기준으로 문자열을 분리 리스트로 저장
* join() : 앞 쪽에 지정한 형식을 포함해서 문자열을 합쳐준다
* strip() : 문자열 양쪽 끝에 공백이나 특수문자를 제거
  * 단점 : 문자열 내부의 특수문자 및 공백을 제거하지 못함
* replace() : 문자열에서 원하는 값을 찾아 다른 값으로 변경

## Strip() : 문자열 양쪽 끝의 공백, 특수문자를 제거 , 그러나 문자열 내부의 공백은 처리하지 못함


```python
h12 = '     공백이 있는 문자열      '
h12
```

## strip의 기능 2가지
* lstrip : 왼쪽 공백을 없애는 함수
* rstrip : 오른쪽 공백을 없애는 함수

```python
h12.lstrip()
```

```python
h12.rstrip()
```

```python
h12.strip() # 양쪽 공백 문자열 삭제
```

```python
h13 = '         $*@&#(알리오알리오*!@&*($%@!$)).... %            '
```

```python
h13.strip('$.()@&.')
```

```python
h13.lstrip('&,@')
```

```python
h13.strip('!.$.&.#')
```

```python
h13= h13.strip(' *@&#().%')
```

```python
h13.replace('.','').replace('!','').replace('$','')
h13
```
