# 1. 두 개의 숫자를 입력받아 합/차/곱/나눗셈을 출력하는 calculation 함수를 작성하라.



```
calculation(3, 4)
```





```
3 + 4 = 7
3 - 4 = -1
3 * 4 = 12
3 / 4 = 0.75
```



````python
def calculation(a,b):
  print(f'{a} + {b} = {a+b}')
  print(f'{a} - {b} = {a-b}')
  print(f'{a} * {b} = {a*b}')
  print(f'{a} / {b} = {a/b}')
````

````python
calculation(3, 4)
````

# 2. 입력된 문자열을 역순으로 출력하는 print_reverse 함수를 정의하라.



```
print_reverse("Hello, World!")
```



```
!dlroW ,olleH
```

````python
def print_reverse(ch):
  return ch[::-1]
````

````python
print_reverse("Hello, World!")
````

# 3. 하나의 리스트를 입력받아 짝수만 화면에 출력하는 print_even 함수를 정의하라.



```
print_even ([1, 3, 2, 10, 12, 11, 15])
```





```
2
10
12
```



````python
def print_even(*args):
  for i in args:
    if i % 2 == 0 :
      print(i)
````

````python
print_even (*[1, 3, 2, 10, 12, 11, 15])
````

````python
list(filter(lambda x: x % 2 ==0,[1,3,2,10,12,15]))
````

# 4. 하나의 딕셔너리를 입력받아 딕셔너리의 key 값을 화면에 출력하는 print_keys 함수를 정의하라.



```
print_keys ({"이름":"김말똥", "나이":30, "성별":0})
```





```
이름
나이
성별
```



````python
def print_keys(**kwargs): #kwargs(keyword arguments)
  print(kwargs)
  # print(type(kwargs))
  for kw,arg in kwargs.items():
    print(kw,':',arg,sep='')
````

````python
print_keys(이름='홍길동',주소='서초구',나이=23)
````

````python
def print_keys(x):
  for i in x.items():
    print(i)
````

````python
print_keys({'이름':'김말똥','나이':30,'성별':0})
````

# 6. 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.



```
pickup_even([3, 4, 5, 6, 7, 8])
```





```
[4, 6, 8]
```



````python
result = []

def pickup_even(*args):
  for i in args:
    if i % 2 == 0:
      result.append(i)
  return result
````

````python
pickup_even(*[3, 4, 5, 6, 7, 8])
````

````python
list(filter(lambda x : x % 2 == 0,[3,4,5,6,7,8]))
````

# 7. 콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하라.



```
convert_int("1,234,567")
```





```
1234567
```



````python
def convert_int(ch):
  covert = ch.replace(',','')
  return int(covert)

````

````python
convert_int("1,234,567")
````

````python
(lambda x: int(x.replace(',','')))("1,234,567")
````

````python
def convert_int(x):
  result = x.replace(',','')
  return int(result)
````
