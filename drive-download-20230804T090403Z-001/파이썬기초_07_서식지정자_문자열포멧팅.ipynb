## 서식 지정자나 format 메서드를 사용하는 이유 :
* 문자열로만 입력 해야 하는 상황에서의 같은 위치의 문자를 여러 번 바꿔야 할 때가 있기 때문이다.


### 1) 문자를 입력할 때 %s

```python
print('이 수업의 이름은 %s 입니다.'%'머신러닝 빅데이터 분석')
```

```python
print('이 수업의 이름은 %s 입니다.'%'파이썬')
```

```python
subject = ['국어','영어','수학','과학','파이썬','장고']
```

```python
for i in subject:
  print('이 수업의 이름은 %s 입니다'%i)
```

### 2) 정수를 입력할 때 %d를 입력합니다


```python
print('올해는 %d 입니다'%2023)
```

```python
year = [2021,2022,2023]

for i in year:
  print('올해는 %d년 입니다'%i)
```

### 3) 실수를 입력할 때 %f

```python
print('사람의 정상체온은 약 %f 도 입니다'%2)
```

### %f는 기본적으로 소수점 6자리까지 표시하므로 f앞에 .자릿수를 지정하면 지정한 자릿수 만큼만 출력된다.

```python
print('사람의 정상 체온은 약 %.1f도 입니다'%36.5)
```

### 4) 서식지정자로 문자열 정렬하기 %길이 s

```python
print('%10s'%'오른쪽정렬')
```

```python
print('%10d'%19500)
```

### 숫자 오른쪽 정렬로 출력하기

```python
print('가격 %10d'%195000)
print('가격 %10d'%71800)
```

```python
print('금일 삼성전자 주가는 %10d원입니다.'%717800)
```

### 여러 항목 동시에 넣기


```python
print('오늘은 %d년 %d월 %d일 입니다'%(2023,10,20))
```

```python
company = ['삼성전자','LG전자','현대차','대한항공','하나투어']
stock = [71800,121000,195000,31900,86600]

for i,j in zip(company,stock):
  print('%8s의 주가는 %10d 입니다'%(i,j))
```

### 2. format 메서드 사용하기
* print({인덱스}'.format(값))
* print({인덱스0}{인덱스1}{인덱스2}'.format(값1,값2,값3))


```python
print('{0}의 주가는 {1}원 입니다'.format('삼성전자',78000))
print('{1}의 주가는 {0}원 입니다'.format('삼성전자',78000))
```

```python
intList = [1, 2, 3, 4, 5]
strList = ["reakwon", "tistory", "com"]

for i,j in zip(intList,strList):
  print(f'{i}와 {j}')
```

### format 메서드로 같은 값을 여러개 넣기

```python
print('{0}의 주가는 {1}원 입니다.{0}의 상한가는 {2},\n {0}의 하한가는 {3}입니다'.format('삼성전자',78000,82000,75000))
```

### format 메서드에 변수로 값 넣기



```python
drama = '오징어게임'
rank = 1
```

```python
print('넷플릭스 {}위 드라마는 {}입니다'.format(rank,drama))
```

```python
print(f'넷플릭스{rank}위 드라마는 {drama}입니다')
```

```python
fruit = ['사과','바나나','야채','토마토','자두']

for i in enumerate(fruit):
  print(f'{i[1]}의 순서는 {i[0]}')
```

### 5. format 메서드로 문자열 정렬하기

* print('{인덱스 : <길이}'.format(값))
* print('{인덱스 : 방향  <길이}'.format(값))

```python
# 오른쪽 방향으로 공백 포함 10칸
print('{0:<10}'.format('python                 '))
```

```python
# 왼쪽방향 공백 포함 10칸
print('{0:>10}'.format('python'))
```

### 1) 숫자 개수 통일하기

* print('{인덱스:채울숫자 길이 서식지정자}'.format(값))
* 단 서식 지정자는 %는 빼고 적는다

```python
print('{0:02d}'.format(1))
```

```python
# 실수로 채울경우
print('{0:03.2f}'.format(1))
```

```python
intList = [1, 2, 3, 4, 5]
print('{0:0.2f}'.format(intList))
```

```python
month = range(1,13)
day = range(1,32)


for m in month:
    for d in day:
        print('{0:02d}월 {1:02d}일'.format(m,d))


# for m,d in zip(month,day):
#   print('{0:002d}월 {}일'.format(m,d))


```

```python
print('{0:2d}'.format(day))
```
