## 반복문 연습문제 1

리스트에 동물 이름 저장돼 있다.

리스트 = ['dog', 'cat', 'parrot']

## 반복문 연습문제 5

표준 입력으로 나이(만 나이)가 입력받아(입력 값은 7 이상 입력됨). <br>
교통카드 시스템에서 각 나이에 맞게 요금을 차감한 뒤 잔액이 출력되게 만드세요
(if, elif 사용). <br>
현재 교통카드에는 10,000원이 들어있습니다.<br>
while 반복문을 사용해서 잔액이 부족해질 때까지 반복해서 나이를 입력하고<br>
잔액이 450원 미만이 되면 '잔액이 부족합니다. 카드를 충전해 주세요'를 출력하세요.<br>

** 요금 **<br/>
어린이(초등학생, 만 7세 이상 12세 이하): 450원<br/>
청소년(중∙고등학생, 만 13세 이상 18세 이하): 720원<br/>
어른(일반, 만 19세 이상): 1,250원

````python

money = 10000

while True :
  age = int(input('나이 입력 >> '))
  if 7 <= age <=99:
    if 7 <= age <= 12:
      money -= 450
    elif 13 <= age <= 18:
      money -=720
    elif age >=19 :
      money -= 1250
    if money <= 0:
      print('잔액이 부족합니다 충전해주세요')
      break
    else:
      print(f'남은금액 {money}원')
  else:
    print('나이는 99세까지입니다')
````
