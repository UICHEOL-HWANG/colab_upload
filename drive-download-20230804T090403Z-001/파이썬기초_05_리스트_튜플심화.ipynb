## 리스트와 튜플, 문자열의 공통점
* 순서가 있고 연속된 값이 들어있다.
* 이러한 연속적인 자료형을 시퀀스 자료형이라고 한다. (시퀀스 객체라고도 함)

* 시퀀스 객체 : list , tuple , string , range() 자료형


## 1. 특정 값이 스퀀스 객체 내에 있는지 확인하기 .
: in 시퀀스 객체 (list,tuple,string,range())

```python
a = list(range(1,10))
a
```

```python
7 in a
```

```python
s = '우리는 오늘 삼복 더위에서 파이썬을 공부 하고 있습니다.'
```

```python
'삼복' in s
```

## 2. 특정 값이 없는지 확인 하기
* 값 not in 시퀀스 객체

```python
'삼계탕' not in s
```

## 3. 시퀀스 객체 연결하기
* '+' 연산자로 연결이 가능하다.
* 단 , range는 불가능 하다.
* list 함수중 extend와 동일

```python
a = list(range(10))
b = list(range(10,20))
print(a,b)
```

```python
# a,b를 더하려면 원래는 extend를 써야하지만
c = a+b
```

```python
c
```
